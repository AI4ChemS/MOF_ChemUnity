```python
df = pd.read_csv('Untracked/full_graph/water_stability_v1.csv')


# 1. Drop the first unnamed (empty) column
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# 2. Remove rows where 'Value' is "Not Provided"
df = df[df['Value'] != "Not Provided"]
df = df[df['Value'] != "Not provided"]
df = df[df['Value'] != "Not applicable"]

# Optional: Save the cleaned DataFrame
df.to_csv("cleaned_file.csv", index=False)
```
